# Financial Data 표현: chartSeries

```r
B = new.env()
getSymbols('BMW.DE', env=B, from='2010-01-01', to='2013-12-31')
BMW = B$BMW.DE
head(BMW)
```

```r
chartSeries(BMW, multi.col=T, theme='white')
addBBands()
addMACD()
```

```r
B.Return = log(BMW$BMW.DE.Close/BMW$BMW.DE.Open)
qqnorm(B.Return, main='Normal Probability Plot of BMW daily Log Returns', xlab='Theoretical Quantiles', ylab='Sample Quantiles')
qqline(B.Return, col='red', lwd=2.5)
```

# Big Data Analysis using R: bigmemory

```r
install.packages('bigmemory')
install.packages('biganalytics')
```

```r
library('bigmemory')
library('biganalytics')
```

```r
x = read.big.matrix('bmw.csv', header=T, type='double')
x = x[, -1]
class(x)
```

```r
Results = lapply(1:10, function(i) {bigkmeans(x, center=i, iter.max=100, nstart=10)})
Within = lapply(Results, function(x) sum(x$withinss)/(10^15))
plot(1:10, Within, type='b', xlab='Number of Clusters', ylab='Within Sum of Squares')
```

# K-means Clustering Analysis with Big Data

```r
R3 = bigkmeans(x, center=3, iter.max=100, nstart=10)
summary(R3)
```
